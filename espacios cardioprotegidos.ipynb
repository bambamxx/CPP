## Ciencia de datos
</div>

Con Colab, puedes aprovechar por completo las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código que se incluye a continuación usa <strong>NumPy</strong> para generar algunos datos aleatorios y <strong>matplotlib</strong> para visualizarlos. Para editar el código, haz clic en la celda y comienza a editar.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import simpy
import random
import matplotlib.pyplot as plt

# Constantes
SIM_TIME = 24 * 60 # tiempo de simulación en minutos (24 horas)
NUM_PERSONAS = 1000 # número de personas en el edificio
PROB_EVENTO_CARDIO = 0.01 # probabilidad de un evento cardiovascular
PROB_ATENCION_RAPIDA = 0.90 # probabilidad de atención dentro de los primeros 4 minutos en un edificio cardioprotegido

# Edificio Cardioprotegido
class EdificioCardioprotegido:
    def __init__(self, env):
        self.env = env
        self.muertes = 0
        self.action = env.process(self.run())

    def run(self):
        while True:
            yield self.env.timeout(1)  # cada minuto
            for i in range(NUM_PERSONAS):
                if random.random() < PROB_EVENTO_CARDIO:
                    if random.random() < PROB_ATENCION_RAPIDA:
                        tiempo_atencion = random.uniform(2, 4)  # tiempo de atención
                    else:
                        tiempo_atencion = random.uniform(4, 8)  # tiempo de atención
                    if tiempo_atencion > 4:
                        self.muertes += 1  # la persona muere si no se atiende dentro de los 4 minutos

# Edificio No Cardioprotegido
class EdificioNoCardioprotegido:
    def __init__(self, env):
        self.env = env
        self.muertes = 0
        self.action = env.process(self.run())

    def run(self):
        while True:
            yield self.env.timeout(1)  # cada minuto
            for i in range(NUM_PERSONAS):
                if random.random() < PROB_EVENTO_CARDIO:
                    tiempo_atencion = random.uniform(10, 40)  # tiempo de atención
                    if tiempo_atencion > 10:
                        self.muertes += 1  # la persona muere si no se atiende dentro de los 10 minutos

# Simulación
def simulacion(num_simulaciones):
    muertes_cardioprotegido = []
    muertes_no_cardioprotegido = []

    for _ in range(num_simulaciones):
        env = simpy.Environment()

        edificio1 = EdificioCardioprotegido(env)
        edificio2 = EdificioNoCardioprotegido(env)

        env.run(until=SIM_TIME)

        muertes_cardioprotegido.append(edificio1.muertes)
        muertes_no_cardioprotegido.append(edificio2.muertes)

    # Gráfico
    plt.plot(muertes_cardioprotegido, label='Edificio Cardioprotegido')
    plt.plot(muertes_no_cardioprotegido, label='Edificio No Cardioprotegido')
    plt.legend()
    plt.xlabel('Simulación')
    plt.ylabel('Muertes')
    plt.show()

simulacion(50)
```
